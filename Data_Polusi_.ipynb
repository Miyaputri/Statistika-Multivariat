```python
#Random Generate Data

#Import Library
import numpy as np
import pandas as pd

np.random.seed(42)

# Matriks kovarian yang sama untuk semua kawasan (CO2, NO2, SO2, PM2.5)
cov_matrix = [[5, 1, 0.5, 1],
              [1, 4, 0.8, 0.6],
              [0.5, 0.8, 3, 0.4],
              [1, 0.6, 0.4, 6]]

# Mean untuk setiap kawasan
mean_industri = [10, 8, 5, 12]
mean_permukiman = [8, 6, 4, 10]
mean_hutan = [6, 4, 3, 7]

# Jumlah sampel per kawasan
n = 50

# Generate data multivariat normal
data_industri = np.random.multivariate_normal(mean_industri, cov_matrix, size=n)
data_permukiman = np.random.multivariate_normal(mean_permukiman, cov_matrix, size=n)
data_hutan = np.random.multivariate_normal(mean_hutan, cov_matrix, size=n)

# Gabungkan ke DataFrame
df = pd.DataFrame(
    np.vstack([data_industri, data_permukiman, data_hutan]),
    columns=['CO2', 'NO2', 'SO2', 'PM25']
)
df['kawasan'] = ['Industri'] * n + ['Permukiman'] * n + ['Hutan'] * n

# Lihat sebagian data
print(df.head())
```

```python
from google.colab import files
df.to_csv('data_polusi.csv', index=False)
files.download('data_polusi.csv')
```
